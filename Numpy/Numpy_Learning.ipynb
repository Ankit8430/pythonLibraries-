```python
# Array Stacking and splitting
```

```python
a=np.array([1,2,3])
b=np.array([4,5,6])

print(np.vstack((a,b))) # Vertical stacking -row wise
print(np.hstack((a,b))) # Horizontal Stacking -column wise
```

```python
c=np.array([[1,2,3],[4,5,6]])
print(c)
```

```python
split=np.hsplit(c,3) #horizontal split
for s in split:
    print(s)
```

```python
split=np.vsplit(c,2)#Vertical Split
for h in split:
    print(h)
```
